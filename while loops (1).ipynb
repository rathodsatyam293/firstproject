```python
# To count if there is no duplicates in your search
# Two Pointer Approach

s = "RAJASAM"

size = len(s)
start = 0
current = 1
count = 0
b =True
while start < size-1:
    print(s[start],start, s[current],current,"Count :-" , count)

    if s[start] == s[current]:
        b = False
    current+=1
    if current == size:
        if b == True:
            count+=1
        print("------RESET-----",count)
        start+=1
        current = start+1
        b=True
```

```python
#check palindrome number
n = 121
s =str(n)
left = 0
right = len(s)-1

is_pal = True

while left<right:
    if s[left]!=s[right]:      # l-r and r-r then l-a and r-a then l-c and r-c
        is_pal = False
        break
    left+=1
    right-=1
print("Palindrome" if is_pal else "Not a Palindrome")
```

```python
# Reverse string word
sentence = "hello world"

i = 0
length = len(sentence)
result = ""

while i < length:
    while i < length and sentence[i] == " ":
        result += " "
        i += 1
    
    start = i
    while i < length and sentence[i] != " ":
        i += 1
    end = i - 1
    
    while end >= start:
        result += sentence[end]
        end -= 1

print(result)
```

```python
#check armstrong number
n = 153
original = n
total = 0
while n>0:
    digit = n%10
    total = total+digit**3
    n=n//10
print("Armstrong" if total==original else "Not a Armstrong")
```

```python
# Perfect Number

n = 28    
i = 1
total = 0

while i<n:
    if n%i==0:
        total+=i
    i+=1
print(total)
print("Perfect" if total==n else "Not Prefect")
```

```python
# Count uppercase and lowercase letters in a string
word = "PyTHon"

i = 0
upper = 0
lower = 0

while i < len(word):
    ch = word[i]
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1
    i += 1

print("Upper:", upper, "Lower:", lower)
```

```python
# Remove a duplicate character
sentence = "hello"
remove_char = "l"

i = 0
result = ""

while i < len(sentence):
    if sentence[i] != remove_char:
        result += sentence[i]
    i += 1

print("After removal:", result)
```
