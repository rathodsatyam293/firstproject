```python
# 1. armstrong number
n = int(input("enter a number:"))
original = n
total = 0
while n > 0:
    digit = n % 10
    total+=digit**3
    n = n//10
print("armstrong" if total==original else "not armstrong")
```

```python
# 3. perfact number
n = int(input("enter a number:"))
i = 1
total = 0
while i < n:
    if n % i == 0:
        total += i
    i += 1
print('Perfect' if total == n else 'Not Perfect')
```

```python
#fibonacci series up to n terms
n = 7
a,b =0,1
count = 0
while count<n:
    print(a,end=" ")
    a,b=b,a+b
    count+=1
```

```python
n = int(input("enter a number:"))
original = n
rev = 0
while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10
print('Palindrome' if original == rev else 'Not Palindrome')
```

```python
# check palindrome number 
word = input("enter a word:")
left = 0
right=len(word)-1
i = 0
is_pal =True

while left<right:
    if word[left]!=word[right]:
        is_pal= False
        break
    left+=1
    right-=1
print('Palindrome' if is_pal else 'Not Palindrome')
```

```python
# GCD of two numbers using while loop
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b != 0:
    temp = b
    b = a % b 
    a = temp   

print("GCD =", a)
```

```python
# find all prime number
n = 12
d = 2
while n > 1:
    while n % d == 0:
        print(d, end=' ')
        n = n // d
    d += 1
```

```python
# normal tringle pattern
for i in range(1,6):
    for j in range(1,i+1):
        print("*",end=' ')
    print()
```

```python
# inverter tringle 
for i in range(6,0,-1):
    for j in range(i):
        print("*",end=' ')
    print()
```

```python
# Prime numbers from 7 to 67 
start = 7
end = 67
num = start

while num <= end:
    if num > 1:
        i = 2
        while i < num:
            if num % i == 0:
                break
            i += 1
        else:
            print(num, end=" ")
    num += 1
```
